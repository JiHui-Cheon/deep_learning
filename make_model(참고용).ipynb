```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import tensorflow as tf
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

df = pd.read_csv('ThoraricSurgery.csv')
df_1 = np.loadtxt('ThoraricSurgery.csv', delimiter=",")
train , test = train_test_split(df, test_size= 0.3)
# np.random.seed(3)
# tf.random.set_seed(3)

train_arr = train.to_numpy()
test_arr = test.to_numpy()
Data_set = np.loadtxt('ThoraricSurgery.csv', delimiter=",")

X = Data_set[:,0:17]
Y = Data_set[:,17]

X_train = Data_set[:,0:17]
# #딥러닝 구조를 결정. 모델을 설정하고 실행하는 부분, 히든레이어 추가
model = Sequential()
model.add(Dense(20, activation='softmax'))
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# print(type(train_arr))
# print(X_train)
# print(train.shape)
```

```python
#딥러닝을 실행
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
a = model.fit(X, Y, epochs=100 , batch_size=20)
```

```python
X_test = test_arr[:,0:17]
Y_test = test_arr[:,17]
score = model.evaluate(X_test, Y_test, verbose=0)
```

```python
X_test = test_arr[:,0:17]
prediction = model.predict(X_test)
human = prediction[int(input('input number : '))]

print(f'환자의 생존률은 {human*100}%, 정확도는 : {score[1]*100}%')
```
